
## üìö  <b> Practice 8. </b> 
Create two random 1-D arrays of length 10. Merge them into a 2x10 array and then a 10x2 array.


```python
# 2 random 1-D arrays
j = np.random.rand(10)
k = np.random.rand(10)

# Concat as 2x10 array
vert = np.vstack((j,k))
# Concat as 10x2 array
horiz = vert.T # or np.column_stack((j,k)).shape
```
